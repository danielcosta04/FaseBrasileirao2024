```python
import pandas as pd

df = pd.read_excel('brasileirao.xlsx')
df.head()
```

###Exercício: análise de Desempenho de Clubes no Campeonato Brasileiro 2024
'Objetivo: '
Os alunos irão aprender a:
Manipular dados de um capeonato de futebol
Calcular e visializar estatísticas de desempenho dos clubes.

###Passo 1: Leitura da Planilha
Carregar os dados do arquivo brasileirao.slsx usando pandas.

###Passo 2: Calcular o saldo de gols (SG)###
Adicionar uma nova coluna chamada "Saldo de Gols" que será a doferemã entre Gols (PRO) e gols Contras(GC).

```python
df['Saldo de Gols'] = df['GP'] - df['GC']
df.head()
```

Passo 3: Calcular a pontuação total

Criar uma coluna chamada "Pontos" onde a pontuação será calculada da seguinte forma:

Vitória (V) = 3 pontos
Empate (E) = 1 ponto
Derrota (D) = 0 ponto

```python
df['Pontos'] = df['V'] * 3 + df['E']
df.head(20)
```

###passo 4: filtrar os clubes com mais de 40 pontos
Filtrar os clubes que têm mais de 40 pontos

```python
clubes_40_pontos = df[df['Pontos'] > 40]
clubes_40_pontos
```

Passo 5: Viadualizar os dados
criar um gráfico de barras para vusualizar a pontuação dos clubes.

```python
df.plot(x='Clube', y='Pontos', kind='bar', title='Pontuação dos Clubes no Brasileirão 2024')
```
